```python
# Combining the data 
final_df=pd.concat([df_train,df_test],axis=0)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
datetimeconversion=pd.to_datetime(final_df["Date_of_Journey"])
final_df["Month"]=datetimeconversion.dt.month
final_df["Day"]=datetimeconversion.dt.day      
final_df["Year"]=datetimeconversion.dt.year

```

```python
final_df.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
arrivaltime=pd.to_datetime(final_df["Arrival_Time"])
final_df["Arrival_hour"]=arrivaltime.dt.hour
final_df["Arrival_minute"]=arrivaltime.dt.minute
```

```python
final_df.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
deptime=pd.to_datetime(final_df["Dep_Time"])
final_df["Dep_hour"]=deptime.dt.hour
final_df["Dep_minute"]=deptime.dt.minute
```

```python
final_df.drop(["Dep_Time"],axis=1,inplace=True)
```

```python
final_df.head() 
```

```python
final_df.info() 
```

```python
final_df["Total_Stops"]=final_df["Total_Stops"].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
final_df["Total_Stops"].fillna(1,inplace=True)  
```

```python
final_df["Total_Stops"].isnull().sum()
```

```python
final_df.head()
```

```python
final_df.drop("Route",axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df["Additional_Info"].unique()
```

```python
# Converting duration into minutes
final_df["Duration"]=final_df["Duration"].str.replace("h","*60").str.replace(" ","+").str.replace("m","*1").apply(eval)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df["Airline"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
final_df["Airline"]=encoder.fit_transform(final_df["Airline"])
final_df["Source"]=encoder.fit_transform(final_df["Source"])
final_df["Destination"]=encoder.fit_transform(final_df["Destination"])
final_df["Additional_Info"]=encoder.fit_transform(final_df["Additional_Info"])
```
